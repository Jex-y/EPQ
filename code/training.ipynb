```python
import yolo
```

```python
train_dataset = yolo.WIDERDataset("data/WIDER/annotations/train.txt", image_size=416, augment=False)
val_dataset = yolo.WIDERDataset("data/WIDER/annotations/val.txt", image_size=416, augment=False)
```

```python
model = yolo.models.YOLOv4(num_classes=1)
```

```python
model.fit(train_dataset, val_dataset=val_dataset,  epochs=100, start_lr = 5e-5, end_lr = 1e-9)
```
